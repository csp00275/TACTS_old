```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from matplotlib.patheffects import withStroke


# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (8,2)
plt.rcParams['font.size'] = 12

# 2. 필요한 변수 지정

D = 200 
d = 100
R = D/2
r = d/2
L = 160
alpha = np.deg2rad(27)

nx = 12
ny = 2

ns = nx*(ny)

cp = 80 #center position
op = 40 #offset position

# 3. 그래프 그리기
fig, ax = plt.subplots()

ax.set_xlim(0, 630)
ax.set_ylim(0, 160)
ax.set_xlabel('width')
ax.set_ylabel('height')


# 4. 사각형 그리기
ax.add_patch(
   patches.Rectangle(
      (0, 0),                   # (x, y)
      2*np.pi*R, L,                     # width, height
      edgecolor = 'lightblue',
      facecolor = 'lightblue',
      fill=True,
   ))
for j in range(0,ny):
    for i in range(0,nx):
        ax.add_patch(
       patches.Ellipse(
          (2*np.pi*R/nx*(i+1/2), cp+(1-2*j)*op),                   # (x, y)
          (R-r)/np.cos(alpha/2)*alpha, (R-r)*2*np.tan(alpha/2),                     # width, height
          edgecolor = 'red',
          facecolor = 'red',
          fill=True,
       ))

        # 5. 빨간 원 사이 거리 표시
        c1 = (2*np.pi*R/nx*(1/2), cp+1*op)
        c2 = (2*np.pi*R/nx*(3/2), cp+1*op)
        c3 = (2*np.pi*R/nx*(1/2), cp+-1*op)

        d_hor = np.sqrt((c2[0]-c1[0])**2 + (c2[1]-c1[1])**2)
        d_dia = np.sqrt((c3[0]-c1[0])**2 + (c3[1]-c1[1])**2)

        text_hori = ax.annotate(str(round(d_hor, 2)), ((c1[0]+c2[0])/2, (c1[1]+c2[1])/2), 
                    xytext=(0, 10), textcoords='offset points', ha='center', va='bottom', color='black')
        text_hori.set_path_effects([withStroke(linewidth=2, foreground='white')]) 

        ax.plot([c1[0], c2[0]], [c1[1], c2[1]], linestyle='--', color='black')




        text_dia = ax.annotate(str(round(d_dia, 2)), ((c1[0]+c3[0])/2, (c1[1]+c3[1])/2), 
         xytext=(-10, -10), textcoords='offset points', ha='center', va='bottom', color='black',
         rotation=-90)

        text_dia.set_path_effects([withStroke(linewidth=2, foreground='white')]) 


        ax.plot([c1[0], c3[0]], [c1[1], c3[1]], linestyle='--', color='black')




        # Calculate the angle between the black dot line
        angle = np.rad2deg(np.arctan2(c2[1]-c1[1], c2[0]-c1[0]) - np.arctan2(c3[1]-c1[1], c3[0]-c1[0]))
        if angle < 0:
            angle += 360

            # Limit the angle to a maximum of 90 degrees
        if angle > 90:
            angle = 180 - angle

        # Add the angle arc to the plot
        arc_center = c1
        arc_radius = 20
        arc_start_angle = np.rad2deg(np.arctan2(c3[1]-c1[1], c3[0]-c1[0])) + 5 
        arc_end_angle = np.rad2deg(np.arctan2(c2[1]-c1[1], c2[0]-c1[0])) - 5
        ax.add_patch(patches.Arc(arc_center, arc_radius*2, arc_radius*2, 0, arc_start_angle, arc_end_angle, edgecolor='black', lw=1.5))

        # Add the angle annotation to the plot
        text = ax.annotate(str(round(angle, 2)) + '$^\circ$', ((c1[0]+c2[0])/2-20, (c1[1]+c2[1])/2+20), 
                     xytext=(0, -30), textcoords='offset points', ha='left', va='bottom', color='black')
        text.set_path_effects([withStroke(linewidth=2, foreground='white')]) 




# 6. 텍스트 삽입하기

plt.title('LABS_v3')

plt.show()

print('number of sensors : {0}'.format(ns))
```
