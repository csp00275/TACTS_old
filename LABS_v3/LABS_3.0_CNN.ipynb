```python
import cv2
import numpy as np

# color 설정

blue_color = (255, 0, 0)
green_color = (0, 255, 0)
red_color = (0, 0, 255)
white_color = (255, 255, 255)

# 모두 0으로 되어 있는 빈 Canvas(검정색)

size = 200

img = np.full((size,size,3),(255,255,255),dtype = np.uint8)

# toDeg

toDeg = 180/np.pi
roRad = 1/toDeg

# Red circle Drawing

num_cir = 12


for i in range(0,num_cir):
    
    theta = i * 360 / num_cir * roRad
    radius = 70
    cos = int(radius * np.cos(theta)+ size/2)
    sin = int(radius * np.sin(theta)+ size/2)
    img = cv2.circle(img, (cos, sin), 5, red_color, -1)
    
    #print(cos,sin)

cv2.imwrite('myCNN_fig/test.jpg',img)
cv2.imshow('circle',img)

cv2.waitKey(0)

cv2.destroyAllWindows()
```
