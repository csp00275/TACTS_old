```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers
import matplotlib.pyplot as plt

tf.random.set_seed(1)
filename = 'D:/ME program/2017-PhD/01_research/02Study/34_ballon_like_cover/DeepLearing/SensorCallbi/forLearing_final.csv'
temp_data = pd.read_csv(filename)

x_data = temp_data.iloc[:,:10]
y_data = temp_data.iloc[:,10:]

model = tf.keras.Sequential()
model.add(layers.Dense(5, activation = 'relu', input_shape = [x_data.shape[1]]))
model.add(layers.Dense(10, activation = 'relu'))
model.add(layers.Dense(16, activation = 'relu'))
model.add(layers.Dense(8, activation = 'relu'))
model.add(layers.Dense(3))

model.compile(loss = 'mse', optimizer = tf.keras.optimizers.Adam(0.00001), metrics = ['mae', 'mse'])

hist = model.fit(x_data, y_data, epochs=10000, verbose=1)

model.save('842_E10000.h5')
```

```python
plt.plot(hist.history['loss'])
plt.ylabel('mse')
plt.xlabel('epochs')
plt.show()
```
